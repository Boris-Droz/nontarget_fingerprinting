```python
#The purpose of this file is to create a filter that only retains non-target features that have quantified values in >half of the samples of a particular source type
import pandas as pd
```

### GW: Retain rows that have values in 8 or more columns of GW subset

```python
#Reading in data
df_gw = pd.read_csv(r'240603-NTA-Normalized-FULL-Input.csv', header=0)

#Define concentration columns
concentration_columns = df_gw.columns[2:17]
min_nonzero_columns = 8
print(concentration_columns)
```

```python
#Creating the filter
filtered_df = df_gw[df_gw[concentration_columns].gt(0).sum(axis=1) >= min_nonzero_columns]
```

```python
print(filtered_df)
```

```python
filtered_df.to_csv('filtered_data_GW_NEW.csv', index=False)
```
